```python
from openai import OpenAI
import os
from dotenv import load_dotenv

load_dotenv()

# 모델 목록 확인
# client = OpenAI()
# models = client.models.list()
# print(models)

# GPT 클라이언트 객체 생성
client = OpenAI(
    api_key=os.getenv("OPENAI_API_KEY")
)

response = client.responses.create(
    model = "gpt-4.1-mini",
    input = "강남구에 가성비 좋은 식당 알려줘"
)

print(response.output_text)
```
