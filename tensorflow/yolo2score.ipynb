```python
# Precision, Recall, mAP нҷ•мқён•ҳкё°

from ultralytics import YOLO

model = YOLO("yolo11n.pt")

metrics = model.val(
    data = r"./aquarium_dataset/data.yaml", imgsz=640
)

print("Precision:", metrics.box.map)            # Precision: 0.011751987471109789
print("Recall:", metrics.box.map50)             # Recall: 0.01988449241985285
print("mAP50:", metrics.box.map50)              # mAP50: 0.01988449241985285
print("mAP50-95:", metrics.box.map)             # mAP50-95: 0.011751987471109789

# м җмҲҳк°Җ лӮ®мқҖ мқҙмң лҠ” лӘЁлҚёмқҙ н•ҷмҠөлҗҳм§Җ м•Ҡм•ҳкё° л•Ңл¬ёмһ…лӢҲлӢӨ. лӘЁлҚёмқ„ н•ҷмҠөмӢңнӮӨл©ҙ м җмҲҳк°Җ н–ҘмғҒлҗ  кІғмһ…лӢҲлӢӨ.
# Backboneмқҙ YOLOv11nмқё лӘЁлҚёмқ„ мӮ¬мҡ©н–Ҳкё° л•Ңл¬ём—җ, н•ҷмҠөмқҙ мҷ„лЈҢлҗҳл©ҙ лҚ” лӮҳмқҖ Precision, Recall, mAPлҘј кё°лҢҖн•  мҲҳ мһҲмҠөлӢҲлӢӨ.
# BackboneмқҖ coco datasetмңјлЎң мӮ¬м „ н•ҷмҠөлҗң лӘЁлҚёмқ„ мӮ¬мҡ©н–Ҳкё° л•Ңл¬ём—җ, н•ҷмҠөмқҙ мҷ„лЈҢлҗҳл©ҙ лҚ” лӮҳмқҖ м„ұлҠҘмқ„ кё°лҢҖн•  мҲҳ мһҲмҠөлӢҲлӢӨ.
```
