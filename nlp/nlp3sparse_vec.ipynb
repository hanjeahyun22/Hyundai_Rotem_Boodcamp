```python
# 희소벡터 표현 방식
# CountVector - 단어 등장 횟수 기반 - 정수로 표현
# TfidfVectorizer : 문서에서 자주 나타나는 단어에 가중치를 높게 주되, 모든 문서에서 자주
#                   등장하는 단어에는 패널티 부여 - 실수로 표현

# CountVector : 단어가 몇 번 나왔는가?
# TfidfVectorizer : 그 단어가 문서에서 얼마나 중요한가?
# Word2Vec : 그 단어가 어떤 의미적 위치에 있는가?

# CountVectorizer와 TfidfVectorizer는 '문장 안에 어떤 단어가 들어 있는가?'를 숫자로 바꾸는 방식이고,
# Word2Vec은 '단어가 어떤 의미적 위치에 있는가'를 숫자로 바꾸는 방식이다.
# 키워드가 정화기히 맞아야하는 검색에는 CountVector, TfidfVectorizer 가 효과적이고
# 의미가 비슷한 단어까지 찾고 싶다면 Word2Vec이 적합하다.
```

```python
print('CountVectorizer 연습')
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer

content = ['How to format my hard disk', 'Hard disk format format problems']

count_vec = CountVectorizer(analyzer='word', min_df=1)
# print(count_vec)
tran = count_vec.fit_transform(content)
print(tran)
print(count_vec.get_feature_names_out())
# ['disk' 'format' 'hard' 'how' 'my' 'problems' 'to']
#    0        1       2     3    4       5        6
print(tran.toarray())
# [[1 1 1 1 1 0 1]
#  [1 2 1 0 0 1 0]]
```

```python
print('TfidfVectorizer 연습')
print(content)
tfidf_vec = TfidfVectorizer(analyzer='word', min_df=1)
# print(tfidf_vec)
tran2 = tfidf_vec.fit_transform(content)
print(tran2)
print(tfidf_vec.get_feature_names_out())
print(tran2.toarray())   # 출현 빈도 확률값 출력
# [[0.33471228 0.33471228 0.33471228 0.47042643 0.47042643 0.  0.47042643]
#  [0.35409974 0.70819948 0.35409974 0.         0.         0.49767483   0.        ]]
# 0.33471228 : 0번째 문장에서 0번째 단어('disk')는 빈도에 대한 0.3347% 정도의 확률값을 가
```
