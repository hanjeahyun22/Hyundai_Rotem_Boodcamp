```python
# LSTM 으로 다음 숫자 예측
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, GRU, Dense, Input

x = np.array([[1,2,3],[2,3,4],[3,4,5],[4,5,6],[5,6,7],[6,7,8],[7,8,9],[8,9,10]])
y = np.array([4,5,6,7,8,9,10,11])

x = x.reshape((8, 3, 1))    # 입력형태(samples, time_steps, features)

model = Sequential()
model.add(Input(shape=(3, 1)))  # 시계열(time_steps) 3개, features=1
model.add(LSTM(32, activation='tanh'))
# model.add(LSTM(32, activation='tanh', return_sequences=True))  # 표준
# model.add(GRU(32, activation='tanh', return_sequences=True))   # 계산량 적고 속도는 빠름
model.add(Dense(16, activation='relu'))
model.add(Dense(1, activation='linear'))
model.summary()
```

```python
model.compile(optimizer='adam', loss='mse')

from tensorflow.keras.callbacks import EarlyStopping
es = EarlyStopping(monitor='loss', patience=3, mode='auto')
model.fit(x, y, epochs=1000, batch_size=1, verbose=2, callbacks=[es])
```

```python
print('예측값 : ', model.predict(x).ravel())
print('실제값 : ', y.ravel())

# 새로운 값으로 예측
x_input = np.array([0, 1, 2])
x_input = x_input.reshape((1,3,1))

new_pred = model.predict(x_input)
print(new_pred.ravel())
```
